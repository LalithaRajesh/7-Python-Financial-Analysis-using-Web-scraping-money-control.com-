```python
import csv
import pandas as pd
import urllib.request as urllib2
from bs4 import BeautifulSoup
import warnings

warnings.simplefilter('ignore',UserWarning)
```

```python
page=urllib2.urlopen("http://www.moneycontrol.com/mutual-funds/performance-tracker/ranks/diversified-equity-2.html")
soup_1 = BeautifulSoup(page)
Id=[]
Name=[]
Links=[]
all_links=soup_1.find_all('a',class_='bl_12')
for i in all_links:
    Id.append((i['href'].split("/")[-1]))
    Links.append(i['href'])
    Name.append(i.find(text=True))
link_list = pd.DataFrame(
    {'Name': Name,
     'Links': Links,
     'Id':Id
     
    })
link_list
```

```python
link_list =link_list[2:4]
```

```python
Holding=[]
# web=['/mutual-funds/nav/reliance-growth-fund-rp/MRC008','/mutual-funds/nav/tata-equity-p-e-fund/MTA125']
all_columns_names=['Fund Class','Name', 'Sector' ,'Qty', '%' ,'Funds_Name' ,'Type','Rating']
l=[]
fail=[]
main_funds=[]
for index_link in link_list['Links']:  
    url="http://www.moneycontrol.com"+index_link #Open the main page of the funds 
    try:
        page=urllib2.urlopen(url) # get the request from the page
        print(url.split("/")[-1])
        df_funds=pd.read_html(url,attrs = {'class': 'tblporhd'})  # read html and find the table class name  
        
        for df_funds_range in range(2):
            temp = df_funds[df_funds_range]
            col_name=[]
            col_value=[]
            col_name.append("ID")
            col_value.append(url.split("/")[-1])
            for i in range(2):
            #     print(df_funds[i].as_matrix(columns=None))
                a=df_funds[i].as_matrix(columns=None)
                for i in range(1,len(a)):
                    for j in range(1,len(a[i])):
                        col_name.append(a[i][0]+"-"+a[0][j])
                        col_value.append(a[i][j])
        l1=pd.DataFrame(col_value,col_name).T
        main_funds.append(l1)
        main_funds

    except ConnectionResetError:
#         print('This url failed: ' + holding_url)
#         fail.append(holding_url)

```
